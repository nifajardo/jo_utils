### Get each sheet headers to ensure that all sheets have the same headers

```python
def get_sheet_headers(file_path):
    xls = pd.ExcelFile(file_path)
    sheet_headers = {}
    
    for sheet_name in xls.sheet_names:
        # Read the first row of each sheet to get headers
        df = pd.read_excel(file_path, sheet_name=sheet_name, nrows=1)
        headers = df.columns.tolist()

        # Store the headers in the dictionary
        sheet_headers[sheet_name] = headers
        
    return sheet_headers
```
